```python
import pandas as pd
import numpy as np
```

```python
data0 = pd.read_csv('./data/daily_sales_data_0.csv')
data0.head()
```

```python
data1 = pd.read_csv('./data/daily_sales_data_1.csv')
data1.head()
```

```python
data2 = pd.read_csv('./data/daily_sales_data_2.csv')
data2.head()

```

```python
data =  [data0 , data1 , data2]
data = pd.concat(data)
data = data.reset_index(drop=True)
data
```

```python
data = data[data['product']=='pink morsel']
data
```

```python
data['price'] = data['price'].replace('[\$,]', '', regex=True).astype(float)
data
```

```python
data['sales'] = data['price'] * data['quantity']
data
```

```python
del data['price']
del data['quantity']
data
```

```python
data = data.reset_index(drop=True)
data.shape
```

```python
del data['product']
```
